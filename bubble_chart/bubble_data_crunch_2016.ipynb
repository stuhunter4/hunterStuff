```python
import pandas as pd
```

```python
election_csv = pd.read_csv("../county_data/resources/nytimes_presidential_elections_2016_results_county.csv")
census_csv = pd.read_csv("../county_data/resources/county_facts_formatted.csv")
```

```python
# list of counties, drop the 'County'
county_list = census_csv["County"].tolist()
for x in range(len(county_list)):
    county_list[x] = county_list[x].replace(" County", "")
print(county_list)
```

```python
census_csv
```

```python
# filter election results for state of California only
ca_election = election_csv.loc[election_csv["State"] == "California"]

# list of votes for Clinton, Trump
election_group = ca_election.groupby(["Vote by county"])
clinton_list = election_group["Clinton"].sum().tolist()
trump_list = election_group["Trump"].sum().tolist()

# use list comprehension to add two lists
total_results = [clinton_list[i] + trump_list[i] for i in range(len(clinton_list))]

# same operation to get average vote for average vote per candidate
clinton_avg_list = [(clinton_list[j] / total_results[j])*100 for j in range(len(clinton_list))]
trump_avg_list = [(trump_list[k] / total_results[k])*100 for k in range(len(trump_list))]

# start with basic dataframe of data you are iterating through
election_df = pd.DataFrame({"Clinton": clinton_avg_list, "Trump": trump_avg_list})

# empty list for our results
election_list = []

# iterate over each row and append
for index, rows in election_df.iterrows():
    if rows["Clinton"] > rows["Trump"]:
        winner = rows["Clinton"] * -1
        election_list.append(winner)
    else:
        winner = rows["Trump"]
        election_list.append(winner)

# list of 2016 election winner per county, to designate colors
person_list = []
for n in range(len(election_list)):
    if election_list[n] > 0:
        person_list.append('rgba(255, 0, 0, 0.6)')
    else:
        person_list.append('rgba(0, 0, 255, 0.6)')
```

```python
print(person_list)
```

```python
pop_list = census_csv["Population"].tolist()
popd_list = census_csv["Population per square mile"].tolist()
ed_list = census_csv["Bachelor's degree or higher, percent of persons age 25+"].tolist()
print(pop_list)
print(popd_list)
print(ed_list)
```

```python
# prepare dataframe for plotly.express chart
df = pd.DataFrame({"Population per square mile": popd_list, "Bachelor's degree or higher, percent of persons age 25+": ed_list,
                     "Population": pop_list, "2016 Winner": person_list, "County": county_list})
```

```python
import json
# use .to_dict and 'records' orient to make our desired list
# can be used as 'data.js', or .insert_many with PyMongo
list_d = df.to_dict(orient='records')
```

```python
# write to a text file for our data.js file
with open("weekly_dict.txt", "w") as file:
    file.write(json.dumps(list_d))
```
