```python
import pandas as pd
```

```python
# read in the csv file
census = pd.read_csv("../county_data/resources/CA_Counties_Census_Data.csv")
laus = pd.read_csv("../county_data/resources/CA_Counties_LAUS_82020.csv")
ed = pd.read_csv("../county_data/resources/CA_Counties_UvR_Ed.csv")
```

```python
# see what we're working with
census
```

```python
# clean it up by droping row by index and resetting the index
nlaus = laus.drop(19)
nlaus = nlaus.reset_index(drop=True)
nlaus
```

```python
ed["Bachelors or higher percent"] = round(ed["Bachelors or Higher"]/ed["Total Age 25+"] * 100, 1)
```

```python
ed_list = ed["Bachelors or higher percent"].tolist()
print(ed_list)
```

```python
nlaus_list = round(nlaus["Unemployment Rate "] * 100, 1).tolist()
print(nlaus_list)
```

```python
# create new table with the data we want to keep, with new columns that include our calculations
df = census[["County", "Population", "Average Household Size", "Total Area", "Totals Households", "Household Income",
             "Per Capita Income", "Poverty Count", "Total With Any Broadband Subscription",
            "Total Has A Computer"]]

df["Unemployment Rate "] = nlaus_list
df["Bachelors or higher percent"] = ed_list

df["Households with a comp %"] = round((df["Total Has A Computer"]/df["Totals Households"]) * 100, 1)
df["Households with bb %"] = round((df["Total With Any Broadband Subscription"]/df["Totals Households"]) * 100, 1)
df["Population Density"] = round(df["Population"]/df["Total Area"], 1)
df["Poverty Rate"] = round((df["Poverty Count"]/df["Population"]) * 100, 1)
df
```

```python
# reorder, then rename columns to fit our web display later
web_df = df[["County", "Population", "Totals Households", "Average Household Size", "Households with a comp %", "Households with bb %",
             "Bachelors or higher percent", "Household Income", "Per Capita Income", "Poverty Rate",
             "Unemployment Rate ", "Population Density", "Total Area"]]

web_df.columns = ["County", "Population", "Households", "Persons per household", "Households with a computer, percent",
                  "Households with a broadband Internet subscription, percent",
                  "Bachelor's degree or higher, percent of persons age 25+", "Median household income (in 2018 dollars)",
                  "Per capita income in 12 months (in 2018 dollars)", "Poverty Rate",
                  "Unemployment Rate (8/1/2020)", "Population per square mile", "Total area in square miles"]

web_df
```

```python
# use melt method and multiple unpivot columns to create dataframe like the one on census.gov/quickfacts
geo_df = pd.melt(web_df, id_vars = ["County"], value_vars = ["Population", "Households", "Persons per household", "Households with a computer, percent",
                  "Households with a broadband Internet subscription, percent",
                  "Bachelor's degree or higher, percent of persons age 25+", "Median household income (in 2018 dollars)",
                  "Per capita income in 12 months (in 2018 dollars)", "Poverty Rate",
                  "Unemployment Rate (8/1/2020)", "Population per square mile", "Total area in square miles"])
melt_df = geo_df.loc[geo_df["County"] == "Alameda County"]
```

```python
# drop the left column with our county name
melt_df = melt_df[["variable", "value"]]

# reset index so it looks better and is easier to read and recall
melt_df = melt_df.reset_index(drop=True)

# rename our columns to fit the style of the census page
melt_df.rename(columns = {"variable": "", "value": "Alameda County"}, inplace=True)
```

```python
# display the results
melt_df
```

```python
geo_df["County"]= geo_df["County"].str.replace(" County", "")
geo_df
```

```python
import json
```

```python
# use .to_dict and 'records' orient to make our desired list
# can be used as 'data.js', or .insert_many with PyMongo
list_d = geo_df.to_dict(orient='records')
```

```python
# write to a text file for our data.js later
with open("cfacts_dict.txt", "w") as file:
    file.write(json.dumps(list_d))
```
