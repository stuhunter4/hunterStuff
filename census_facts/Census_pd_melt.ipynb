```python
import pandas as pd
```

```python
# read in the csv file
census = pd.read_csv("../county_data/resources/CA_Counties_Census_Data.csv")
```

```python
# see what we're working with
census
```

```python
# create new table with the data we want to keep, with new columns that include our calculations
df = census[["County", "Population", "Total Area", "Median Age", "Per Capita Income", "Poverty Count"]]
df["Population Density"] = round(df["Population"]/df["Total Area"], 1)
df["Poverty Rate"] = round((df["Poverty Count"]/df["Population"]) * 100, 1)
df
```

```python
# use melt method and multiple unpivot columns to create dataframe like the one on census.gov/quickfacts
geo_df = pd.melt(df, id_vars = ["County"], value_vars = ["Population", "Population Density", "Total Area"] )
geo_df = geo_df.loc[geo_df["County"] == "Alameda County"]
```

```python
# drop the left column with our county name
geo_df = geo_df[["variable", "value"]]

# reset index so it looks better and is easier to read and recall
geo_df = geo_df.reset_index(drop=True)

# rename our columns to fit the style of the census page
geo_df.rename(columns = {"variable": "", "value": "Alameda County"}, inplace=True)
```

```python
# display the results
geo_df
```

```python
#same operations to create a table for 'Income & Poverty' dataframe
inc_df = pd.melt(df, id_vars = ["County"], value_vars = ["Population", "Per Capita Income", "Poverty Rate"] )
inc_df = inc_df.loc[inc_df["County"] == "Alameda County"]
inc_df = inc_df[["variable", "value"]]
inc_df = inc_df.reset_index(drop=True)
inc_df.rename(columns = {"variable": "", "value": "Alameda County"}, inplace=True)
inc_df
```
