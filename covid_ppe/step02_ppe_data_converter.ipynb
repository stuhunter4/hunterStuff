```python
import pandas as pd
from datetime import date
import json
```

```python
# read in ppe info for past updates
ppe_crunched = pd.read_csv("resources/weeks_01-39.csv")

# read in ppe info from previous ppe_cruncher_updates notebook
weekly_df = pd.read_csv("resources/weekly_df_backup_03_22_2021.csv")
```

## Combining Data Sets.

```python
# create new week number column in weekly_df, for continuous indexing
weekly_df.rename(columns = {'week number':'old week number'}, inplace = True)

# create new column with new week numbers, drop old column
weekly_df["week number"] = weekly_df["old week number"] + 38
weekly_df.drop(['old week number'], axis = 1)
```

```python
print(weekly_df['week number'].unique())
```

```python
# drop the last week in ppe_crunched with no elements to utilize
ppe_crunched = ppe_crunched.loc[ppe_crunched['week number'] < 39]
print(ppe_crunched['week number'].unique())
```

```python
# append new dataframes to previous ppe info, for a complete dataset
new_df = ppe_crunched.append(weekly_df)
```

```python
print(new_df['week number'].unique())
new_df
```

```python
# redundant backup
new_df.to_csv(f"resources/weeks_01-42.csv", index=False, header=True)
```

## Clean up, formatting.

```python
# fix county with string replace
new_df["county"] = new_df["county"].str.replace("[", "")
new_df["county"] = new_df["county"].str.replace("]", "")
new_df["county"] = new_df["county"].str.replace("'", "")
print(new_df["county"].unique())
```

```python
# create list of product families
product_list = ['Face Shields (Disposable)', 'Surgical Masks', 'Cloth Masks', 'Surgical or Examination Gowns',
            'Examination Gloves', 'N-95 Respirators', 'Hand Sanitizers', 'Wipes', 'Other / None of the above',
            'Cleaning Supplies', 'Swabs', 'Viral Testing Media', 'Pharmaceuticals', 'Shoe Covers', 'Test Kits',
            'KN95 Respirators', 'Coveralls (Hospitals or EMS)', 'Medical Equipment', 'Goggles', 'Lab Supplies',
            'Beds/Cots', 'Body Bags', 'Lab Kit', 'Ventilators', 'Personnel', 'Sample Collecting Kits', 'Lab Platform',
            'Bedding Accessories', 'Fit Testing Supplies']
print(product_list)
print(len(product_list))
```

```python
# iterate through product list and string replace each remaining column left to clean
for product in product_list:
    new_df[product] = new_df[product].str.replace("[", "")
    new_df[product] = new_df[product].str.replace("]", "")
```

```python
# replace nan values in dataframe with zeroes, for entire dataframe
new_df = new_df.fillna(0)
```

```python
# drop the last week in ppe_crunched with no elements to utilize
new_df = new_df.loc[new_df['week number'] < 42]
```

```python
# columns with brackets are strings
new_df
```

```python
# formatting
for product in product_list:
    new_df[product] = new_df[product].astype(float)
```

```python
# check our results
new_df
```

```python
# use .to_dict and 'records' orient to make our desired list
list_d = new_df.to_dict(orient='records')
```

```python
# write to a text file for our data.js later
with open("weekly_dict.txt", "w") as file:
    file.write(json.dumps(list_d))
```

```python
# create dataframe of weeks and week numbers, using groupby method
week_group = new_df.groupby(["week"])
weeks = week_group["week number"].unique()
group_df = pd.DataFrame({"week number": weeks})

# reset index, and cleanup for []
group_df.reset_index(inplace=True)
group_df["week number"] = group_df["week number"].str[0]
group_df
```

```python
# export week information to a separate TXT file
list_d = group_df.to_dict(orient='records')
with open("week_data.txt", "w") as file:
    file.write(json.dumps(list_d))
```
